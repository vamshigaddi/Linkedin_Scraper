```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

class LinkedInScraper:
    def __init__(self, username, password):
        chrome_options = Options()
        # chrome_options.add_argument("--headless")
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
        self.login(username, password)

    def login(self, username, password):
        try:
            self.driver.get("https://www.linkedin.com/login")
            WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.NAME, "session_key")))
            self.driver.find_element(By.NAME, "session_key").send_keys(username)
            self.driver.find_element(By.NAME, "session_password").send_keys(password)
            self.driver.find_element(By.XPATH, "//button[@type='submit']").click()
            time.sleep(2)
        except Exception as e:
            print(f"Error during login: {e}")

    def send_connection_request(self, profile_url, message):
        try:
            # Navigate to the profile URL
            self.driver.get(profile_url)
            time.sleep(2)
            
            # Click the "More actions" button
            try:
                # Locate the "More actions" button
                more_actions_button = WebDriverWait(self.driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//button[@aria-label='More actions' and contains(@class, 'artdeco-dropdown__trigger')]"))
                )
                
                # Click using JavaScript
                self.driver.execute_script("arguments[0].click();", more_actions_button)
                time.sleep(1)  # Wait for the dropdown menu to open
                print("Successfully clicked the More actions button with JavaScript.")
            except Exception as e:
                print(f"Error clicking More actions button: {e}")
                        
            # Click the "Connect" button in the dropdown
            try:
                # Locate the Connect button
        
                connect_button = WebDriverWait(self.driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'artdeco-dropdown__item') and @role='button' and contains(@aria-label, 'Invite')]"))
                )
            
                # Click the button using JavaScript
                self.driver.execute_script("arguments[0].click();", connect_button)
                time.sleep(1)
                print("Successfully clicked the Connect button with JavaScript.")
            except Exception as e:
                print(f"Error clicking Connect button: {e}")

            # Click the "Add a note" button
            try:
                add_note_button = WebDriverWait(self.driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Add a note']"))
                )
                add_note_button.click()
                time.sleep(1)
            except Exception as e:
                print(f"Error clicking Add a note button: {e}")
                return

            # Enter the message in the note input field
            try:
                message_input = WebDriverWait(self.driver, 10).until(
                    EC.visibility_of_element_located((By.ID, "custom-message"))
                )
                message_input.send_keys(message)
            except Exception as e:
                print(f"Error entering message: {e}")
                return

            # Click the "Send" button
            try:
                send_button = WebDriverWait(self.driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Send invitation']"))
                )
                if "disabled" not in send_button.get_attribute("class"):
                    send_button.click()
                    print("Connection request sent successfully!")
                else:
                    print("Send button is disabled.")
            except Exception as e:
                print(f"Error clicking Send button: {e}")

        except Exception as e:
            print(f"An error occurred while sending the connection request: {e}")
            
    def close(self):
        # Close the browser session
        self.driver.quit()
```
